```python
stocks = pd.DataFrame({'AAPL': apple['Adj Close'], 'MSFT': microsoft['Adj Close'], 'GOOG': google['Adj Close']})
stocks.head()
```

# new stuff learnt today!

```python
stocks.plot(secondary_y = ['AAPL', 'MSFT'], grid = True)
```

```python
stock_returns = stocks.apply(lambda x: x*1./x[0])
stock_returns.head()
```

```python
stock_returns.plot(grid = True).axhline(y = 1, color = 'black', lw = 2)
```

```python
import numpy as np
```

```python
stocks.head()
```

```python
stock_change = stocks.apply(lambda x: np.log(x) - np.log(x.shift(1)))
stock_change.head()
```

```python
stock_change.plot(grid = True).axhline(y = 0, color = 'black', lw = 2)
```

```python
stock_change.skew()
```

```python
stock_change.plot(kind = 'hist', bins = 40, alpha = 0.6)
```
